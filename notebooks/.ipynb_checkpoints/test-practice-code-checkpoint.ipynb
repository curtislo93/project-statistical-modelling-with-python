```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm

# Generate synthetic data
np.random.seed(42)
n_bike_stations = 50

# Generate random latitude and longitude for bike stations
latitude = np.random.uniform(49.2, 49.3, n_bike_stations)
longitude = np.random.uniform(-123.15, -123.10, n_bike_stations)

# Generate random Yelp ratings for the top-rated restaurants near each bike station
yelp_rating = np.random.uniform(3.0, 5.0, n_bike_stations)

# Create a DataFrame with the synthetic data
data = {
    'Latitude': latitude,
    'Longitude': longitude,
    'Yelp Rating': yelp_rating
}
df_synthetic = pd.DataFrame(data)

# Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(longitude, latitude, c=yelp_rating, cmap='viridis', alpha=0.7)
plt.colorbar(label='Yelp Rating')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Scatter Plot of Bike Stations and Top-rated Yelp Restaurants')
plt.grid(True)
plt.show()

# Fit a linear regression model
X = df_synthetic[['Latitude', 'Longitude']]
X = sm.add_constant(X)  # Add a constant for the intercept term
y = df_synthetic['Yelp Rating']

model = sm.OLS(y, X).fit()

# Print the model summary
print(model.summary())
```
